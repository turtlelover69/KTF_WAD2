```python
import pandas as pd
import enchant
```

```python
data = pd.read_csv('ingredient_dataset/newfood.csv')
checker = enchant.Dict("en_US")
```

```python
ingredients_list = []

def check_word(word):
    check = True
    words_list = word.split(' ')
    for single_word in words_list:
        if not(checker.check(single_word)) or single_word.isnumeric():
            check = False 
    return check
    

for index, ingredients in data.iterrows():
    ingredients_split = ingredients['ingredients'].split(',')
    for ingredient in ingredients_split:
        if check_word(ingredient) and ingredient.lower() not in ingredients_list:
            ingredients_list.append(ingredient.lower())

ingredients_list.sort()

            
```
